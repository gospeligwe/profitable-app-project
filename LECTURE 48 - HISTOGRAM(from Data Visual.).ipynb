```python
#Normal Dist
data1 = stats.norm(0,5).rvs(100)

#Two gamma dist. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])

#Box plot both data1 and data2
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.01) 
```

```python
sns.violinplot(data1,inner='stick')
```

# Lecture 52 - Multiple Regression Plot

```python
from io import StringIO
```

```python
tips = """"\
total_bill tip sex smoker day time size
16.29 1.01 Female No Sun Dinner 2
10.34 1.66 Male No Sun Dinner 3
21.01 3.50 Male No Sun Dinner 3
23.68 3.31 Male No Sun Dinner 2
24.59 3.61 Female No Sun Dinner 4\""""
```

```python
dframe = pd.read_csv(StringIO(tips), sep='\s')
```

```python
dframe
```

```python
dframe.head()
```

```python
sns.lmplot('"total_bill','tip',dframe)
```

```python
sns.lmplot('"total_bill','tip',dframe, order=4,
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('"total_bill','tip',dframe,fit_reg=False)
```

```python
dframe.head()
```

```python
dframe['tip_perct']=100*(dframe['tip']/dframe['"total_bill'])
```

```python
dframe.head()
```

```python
sns.lmplot('size','tip_perct',dframe)
```
